# Librería Pandas para Análisis de Datos

## Introducción

Pandas es una librería de Python creada en 2010 que provee estructuras de datos y funciones de alto nivel para trabajar con datos estructurados o tabulares de forma intuitiva y flexible. Gracias a ella, Python se consolidó como un entorno muy productivo para el análisis de datos.

## Objetos principales

* DataFrame → estructura tabular con etiquetas en filas y columnas.

* Series → arreglo unidimensional con etiquetas.

## Características clave

Combina la computación con arrays de NumPy con funcionalidades de bases de datos (SQL) y planillas de cálculo.

Facilita indexar, filtrar, transformar, agrupar y seleccionar subconjuntos de datos.

Ofrece soporte para manipulación, preparación y limpieza de datos, esenciales en análisis.

Tiene un soporte profundo para series temporales, ya que nació para resolver problemas financieros y de análisis de negocio.

## Funcionalidades destacadas


Ejes con etiquetas que evitan errores de alineación de datos.

Operaciones aritméticas que preservan metadatos.

Manejo flexible de datos faltantes.

Operaciones de merge y similares a SQL.

Una misma estructura para datos de series temporales y datos generales.

## Historia

Creado en 2008 por Wes McKinney en AQR Capital Management.

Expandido en 2011-2012 con la ayuda de Adam Klein y Chang She.

Desde 2013 es un proyecto mantenido por la comunidad, con más de 2.000 contribuidores globales.

## Relación con R

El nombre DataFrame viene del objeto equivalente en R (data.frame).

En R está integrado al lenguaje, mientras que en Python fue necesario desarrollarlo como librería.

## Origen del nombre

“Pandas” proviene de panel data (un término econométrico para datos multidimensionales estructurados) y también de Python Data Analysis.
